```python
import pandas as pd
#simple dataframe
data={
    "department":["IT","IT","HR","HR","finance"],
    "name":["A","B","C","D","E"],
    "marks":[80,85,75,80,90]
}
df=pd.DataFrame(data)
print(df)
```

```python
# grouping operation - sum of marks department wise
print("average  marks by department:")
print(df.groupby("department")["marks"].mean())
```

```python
#total
print("\n Total marks by dep :")
print(df.groupby("department")["marks"].sum())
```

```python
#grouping- count
print("\n student count by department :")
print(df.groupby("department")["marks"].count())
```
