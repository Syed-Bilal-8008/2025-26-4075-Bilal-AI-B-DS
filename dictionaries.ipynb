```python
# Creating a dictionary
d = {
    "name": "Bilal",
    "age": 20,
    "course": "BTech"
    }
print("Original dictionary:", d)
```

```python
d = {
    "name": "bilal",
    "age": 20,
    "course": "BTech"
}
# Accessing values
print("Name:", d["name"])
print("Age:", d.get("age"))
```

```python
d = {
    "name": "Bilal",
    "age": 20,
    "course": "BTech"
}
# Adding a new key-value pair
d["college"] = "XYZ University"
print("After adding:", d)
```

```python
d = {
    "name": "Bilal",
    "age": 20,
    "course": "BTech"
}
# Updating value
d["age"] = 21
print("After updating:", d)
```

```python
d = {
    "name": "Bilal",
    "age": 20,
    "course": "BTech"
}
# Removing elements
d.pop("course")
print("After pop:", d)
removed_item = d.popitem()
print("After popitem:", d)
print("Removed item:", removed_item)
```

```python
d = {
    "name": "Bilal",
    "age": 20,
    "course": "BTech"
}
# Keys, values, items
print("Keys:", d.keys())
print("Values:", d.values())
print("Items:", d.items())
```

```python
d = {
    "name": "Bilal",
    "age": 20,
    "course": "BTech"
}
# Length
print("Length of dictionary:", len(d))
```
