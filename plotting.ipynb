```python
import pandas as pd
import matplotlib.pyplot as plt

# DataFrame
data = {
    "Name": ["A", "B", "C", "D"],
    "Age": [20, 21, 19, 22],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 1️ Bar plot: Marks by Name
plt.bar(df["Name"], df["Marks"])
plt.show()
```

```python
# 2️ Line plot: Marks trend
plt.plot(df["Name"], df["Marks"])
plt.show()
```

```python
# 3️ Scatter plot: Age vs Marks
plt.scatter(df["Age"], df["Marks"])
plt.show()
```

```python
# 4 Pie chart: Percentage of total Marks
plt.pie(df["Marks"], labels=df["Name"])
plt.show()
```
